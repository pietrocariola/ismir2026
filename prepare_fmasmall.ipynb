```python
import os
import pandas as pd
import json
from pathlib import Path
```

```python
FMA_META_PATH = "/home/gpu1/datasets/fma/fma_metadata/tracks.csv" #"/.../.../fma_metadata/tracks.csv"
DS_PATH = "/home/gpu1/datasets/fma/fma_small" #"/.../.../fma_small"
FILES_PER_GENRE = 100
GENRES = ['Electronic', 'Experimental', 'Folk', 'Hip-Hop', 'Instrumental', 'Pop', 'Rock']
DS_NAME = "fmasmall"
EXCLUDE_LIST = ["011298.mp3"]
```

```python
tracks = {
    DS_NAME: [],
}

df = pd.read_csv(FMA_META_PATH, header=1)
df = df[df["subset"]=="small"]

if Path("tracks.json").exists():
    with open("tracks.json", "r") as f:
        saved_tracks = json.load(f)

for genre in GENRES:
    g = df[df["genre_top"]==genre]
    for i in range(FILES_PER_GENRE):
        file = str(g.iloc[i, 0]).zfill(6)+".mp3"        
        file_path = os.path.join(DS_PATH, file[:3], file)
        if saved_tracks:
            if DS_NAME in saved_tracks.keys():
                if file_path in saved_tracks.values():
                    continue
        if file not in EXCLUDE_LIST:
            tracks[DS_NAME].append(file_path)

with open("tracks.json", "w") as f:
    json.dump(tracks, f, indent=2)
```
