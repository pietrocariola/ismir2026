```python
import os
import pandas as pd
import json
```

```python
FMA_META_PATH = "/home/gpu1/datasets/fma/fma_metadata/tracks.csv"
FILES_PER_GENRE = 100
GENRES = ['Electronic', 'Experimental', 'Folk', 'Hip-Hop', 'Instrumental', 'Pop', 'Rock']
```

```python
files = []

f_dict = {}
for genre in GENRES:
    f_dict[genre] = []

fma = pd.read_csv(FMA_META_PATH, header=1)
fma = fma[fma["subset"]=="small"]

for genre in GENRES:
    g = fma[fma["genre_top"]==genre]
    for i in range(FILES_PER_GENRE):
        file = str(g.iloc[i, 0]).zfill(6)+".mp3"
        files.append(file)
        f_dict[genre].append(file)

with open("tracks.txt", "w") as f:
    for file in files:
        f.write(str(file) + "\n")

with open("tracks.json", "w") as f:
    json.dump(f_dict, f)
```

```python
fma = pd.read_csv(FMA_META_PATH, header=1)

g = {}

for genre in GENRES:
    l = []
    df = fma[fma["genre_top"]==genre]
    for i in range(len(df)):
        l.append(int(df.iloc[i,0]))
    g[genre] = l

c = 0
with open("tracks.json", "r") as f:
    data = json.load(f)
    for genre in GENRES:
        for track in data[genre]:
            track = int(track.split(".")[0])
            if track not in g[genre]:
                c += 1
if c == 0:
    print("Validation ok!")
```
